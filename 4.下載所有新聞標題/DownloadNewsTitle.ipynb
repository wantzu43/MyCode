```python
url = 'https://tw.yahoo.com/'
import bs4,requests

r = requests.get(url)
print('網頁下載中')
r.raise_for_status()
print('網頁下載完成')
print('='*50)
soup = bs4.BeautifulSoup(r.text,'lxml')
#先找div 再找超連結a
for i in range(5):
    newstags = soup.select('#panel' + str(i) + '-content')
    print('用這個id的分區數目:',len(newstags))
    headnews = newstags[0].find_all('a')
    print('找到幾個新聞:',len(headnews))
    for news in headnews:
        print('新聞:',news.text)
    print('='*50)
```
